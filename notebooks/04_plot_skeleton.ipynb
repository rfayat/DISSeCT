# Plot skeleton for a segment

```python
import pandas as pd
import matplotlib.pyplot as plt
import sys
sys.path.append("..")
from dissect import skeleton 
```

```python
imu = pd.read_pickle("../data_example/example_imu.pickle")
pose = pd.read_pickle("../data_example/example_pose.pickle")
```

```python
pose.head(10)
```

```python
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection="3d")
skeleton.plot_skeleton_gradient(pose.iloc[0], cmap="winter_r")
ax.set_zlim([0, .10])
```

```python
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection="3d")

pose_to_plot = skeleton.center_pose(pose.iloc[1000:1015], how="last")

skeleton.plot_skeleton_gradient_timeseries(pose_to_plot, cmap="winter_r", alpha_vmin=.01)
ax.set_zlim([0, .2])

ax.elev=30
ax.axis("off")
```
